# Step 0: Load data using the CSV file
---
## Import dependencies

```python
import csv 
import cv2 
import numpy as np
import pickle
```

## Load data and combine images
I have recorded **three laps of counter-clockwise driving and one of clockwise**. In the third counter-clockwise lap, i started from another place of the circuit. In all of them I tried to keep the speed around **20-25 MPH** to manage to archive the smoothest possible driving, keeping the car in the **center of the lane** even during curves.

**I will use the three images** the car takes by adding a certain correction to the registered steering angle. This will provide the model with not only more data, but, in particular, with more turning data, which will make it much more capable of recentering the car in the lane and to avoid failures while turning due to a tendence of going straight.

```python
# List of CSV lines
lines = []

# Load data from the CSV (2 laps of normal driving)
with open('./data/driving_log.csv') as csv_file:
    reader = csv.reader(csv_file)
    for line in reader:
        lines.append(line)

# Load images and measurements from the CSV
## Lists to store images and steering angles
images = []
measurements = []

## Correction to be applied to the steering for the side images (will be positive and negative)
steering_correction = 0.2

## For each lien of the CSV file
for line in lines:
    # Read and append center image and steering angle
    image = cv2.imread(line[0])
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    images.append(image)
    measurements.append(float(line[3]))
    
    # Read and append left image and steering angle+correction
    image = cv2.imread(line[1])
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    images.append(image)
    measurements.append(float(line[3])+steering_correction)
    
    # Read and append right image and steering angle-correction
    image = cv2.imread(line[2])
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    images.append(image)
    measurements.append(float(line[3])-steering_correction)

# Convert from list to np array
x_train = np.array(images)
y_train = np.array(measurements)
```

# Step 2: Create the Convolutional Neural Network
Architecture proposed by NVidia in [this paper](https://images.nvidia.com/content/tegra/automotive/images/2016/solutions/pdf/end-to-end-dl-using-px.pdf). This network gets images in the **YUV** color space and consist in 9 layers:
- 1 hardcoded (not learned) **normalization** layer.
- **5 convolutional** layers. The three first with a 2x2 stride and a 5x5 kernel and, the other 2, with no stride and with a 3x3 kernel.
- **3 fully connected** layers

However, I also removed a bit of the top and the bottom of the frames and I used the normal RGB images, even some other color space might be useful for a more real application in which we need to variate the images to simulate darker driving conditions.

---
## Network definition

```python
from keras.models import Sequential, Model
from keras.layers import Flatten, Dense, Lambda, Convolution2D, Cropping2D

def preproc_layers():
    """
    Create a model with the preprocessing for the images embedded on it.
    """
    
    model = Sequential()
    
    # Remove 50 top pixels and 20 down ones. None on the sides.
    model.add(Cropping2D(cropping=((50,20), (0,0)), input_shape=(160,320,3)))
    
    # Normalize data
    model.add(Lambda(lambda x: (x / 255.0) - 0.5))
    
    return model


def nvidia_cnn():
    """
    Create the end-to-end network proposed by NVIDIA on their paper
    """
    
    # Start with the preprocessing layers
    model = preproc_layers()
    
    # Convolutional layers, ReLU-activated 
    model.add(Convolution2D(24, 5,5, subsample=(2,2), activation='relu')) # 5x5 kernel, 2x2 stride
    model.add(Convolution2D(36, 5,5, subsample=(2,2), activation='relu'))
    model.add(Convolution2D(48, 5,5, subsample=(2,2), activation='relu'))
    model.add(Convolution2D(64, 3,3, activation='relu')) # 3x3 kernel, 1x1 stride
    model.add(Convolution2D(64, 3,3, activation='relu'))
    
    # Flatten the 64 feature maps to a 1 1D tensor
    model.add(Flatten())
    
    # Finish with the fully connected layers
    ## NO activation func. bc this is REGRESSION
    ## 1 output bc we want the steering angle
    model.add(Dense(100)) # 100 neurons
    model.add(Dense(50))
    model.add(Dense(10))
    model.add(Dense(1))
    
    return model
```

## Model creation, compilation, training and saving

```python
# Create the model
model = nvidia_cnn()

# Use Mean Square Error loss and ADAM optimizer
# MSE instead of cross-entropy bc this is a regression problem, I am not classifying
model.compile(loss='mse', optimizer='adam')

# Train using 20% of training set for validation and shuffling the data
model.fit(x_train, y_train, epochs=10, validation_split=2.0, shuffle=True)

#history_object = model.fit_generator(train_generator, samples_per_epoch= \
#                 len(train_samples), validation_data=validation_generator, \
#                 nb_val_samples=len(validation_samples), nb_epoch=3, verbose=1)

# Save the model
model.save('model.h5')
```
